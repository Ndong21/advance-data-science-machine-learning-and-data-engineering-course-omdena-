##### Inferential statistics (2 sample t-test)

Hypothesis

- Ho: There is no difference in the mean BMI between males and females

- Ha: There is a difference in the mean BMI between males and females



```python
# subset the data into male and female

male = data[['BMI']][data['Sex'] == 'Male']

female = data[['BMI']][data['Sex'] == 'Female']
```

Significance level

- The significance level for this test is 5% or 0.05

Computing the p-value

```python
# first install scipy
!pip install scipy
```

```python
# compute the p-value using the stats module in scipy
# first import the stats module

from scipy import stats

# compute the p-value

stats.ttest_ind(a=male['BMI'], b=female['BMI'], equal_var = False)
```

Conclusion
- Since the p-value obtained (5.204) is far greater than the significance level (0.05), the results are not statistically significant hence, we fail to reject the null hypothesis. Therefore, there is no mean difference in BMI for both male and females.

Activity: Outlier detection

```python
# calculate 25th percentile of systolic BP
percentile25 = data['Weight(kg)'].quantile(0.25)

# calculate 75th percentile of systolic BP
percentile75 = data['Weight(kg)'].quantile(0.75)

# calculate the iqr
iqr = percentile75 - percentile25

# calculate the upper and lower thresholds beyong which any value is considered a outlier.
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr

# print the lower and upper limits
print(f'''Values below {lower_limit} and above {upper_limit} are outliers''')
```

```python
# print out the rows that contain outliers
outliers = data[(data['Weight(kg)'] < lower_limit) | (data['Weight(kg)'] > upper_limit) ]
outliers
```

```python
# remove the outliers

no_outliers = data[(data['Weight(kg)'] >= lower_limit) & (data['Weight(kg)'] <= upper_limit) ]
no_outliers.shape
```

Methodology and reasoning for this approach

- Outliers are extreme data points that differ significantly from majority of the data points in a data set. One way to identify outliers is to use the interquantile range (iqr). This theory states that values below the 1st quantile by 1.5 * iqr and values above the 3rd quantile by 1.5 * iqr are outliers.

- To achieve this, we need to use the 25th and 75th percentile and calculate the interquartile range,iqr (percentile75 - percentile25). Once we have the iqr, we calculate the lower and upper limit make the range of outliers.
