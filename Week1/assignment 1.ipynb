Sorting the data

- sorting data means displaying it in a particular order e.g descending order, largest to smallest value etc.

- Now, i am going to sort the data from the oldest participant to the youngest participant.

```python
# sort the data using the age variable and assign the results to a new dataframe 'sorted'

sorted = data.sort_values('Age', ascending=False)

#display the sorted dataset
sorted
```

Grouping the data
- When data is grouped under a certain category, aggregate functions like avg(), max() etc can be applied.

- i will group the data under level of education and obtain the maximum age of participants under each category

```python
# select the level of eduation and age variables to include in this new dataframe
# use the .max() to obtain only the maximum number of years per category
# use sort_values() to display the dataframe in ascending order.

max_years = data[['Education level','Age']].groupby('Education level').max().sort_values(by='Age', ascending=True)

# display the dataframe
max_years
```

##### Feature Engineering

- I will create a new feature from the age feature. the new feature is going to be a categorical data that groups the age into groups (10-13, 14-16, 17-19).

- This is an important task in data analysis as we will often be faced with identify trends and patterns across different age groups rather than across the continous age variable. For example, i might want to identify systolic blood pressure and compare these values across participants aged 10-13 yrs vs those aged 14-16 years and so on

```python
# create a new column in our dataset called age_group
# categories age into 3 categories
# i will use the pandas cut()
# 10 to 13 yrs: (9, 13], 9 is excluded and 13 is included
# 14 to 16 yrs: (13, 16], 13 is excluded and 16 is included
# 17 to 19 yrs: (16, 19], 16 is excluded and 19 is included.

data['Age_group'] = pd.cut(x=data['Age'], bins=[9, 13, 16, 19], labels=['10 to 13 yrs','14 to 16 yrs','17 to 19 yrs'])
```

```python
# display the age and age group columns to verify

data[['Age', 'Age_group']]
```
